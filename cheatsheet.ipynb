## Backpropagation

```python
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)
f = 3 * a**3 - b**2
```

$f = 3a^3 - b^2, \qquad \partial f/  \partial a = 9a^2, \qquad \partial f/  \partial b = -2b$

```python
f.backward(gradient=torch.tensor([1, 1]))
```

```python
print(9 * a**2, a.grad)
print(-2 * b, b.grad)
```
